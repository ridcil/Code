```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def cond(r_sheet, thickness):
    return 1 / (r_sheet * thickness * 1e-7)
```

```python
palette = 'cividis_r'
names = ['Sheet Resistance ($\Omega$ / sq)', 'Conductivity (S/cm)', 'Pt']
# sample
na_pt = pd.DataFrame(np.array([2.158, 2.086, 2.256, 2.128]), columns=[names[0]])
na_pt[names[2]] = 'As deposited'
a_pt = pd.DataFrame(np.array([1.658, 1.607, 1.582, 1.653]), columns=[names[0]])
a_pt[names[2]] = 'Annealed'

data = pd.concat([na_pt, a_pt], ignore_index=True)
data[names[1]] = cond(data[names[0]], 100)
fig, ax = plt.subplots(1, 2, dpi = 200, figsize = (15, 5))
sns.barplot(data = data, x = names[2], y = names[0], palette = palette, ax = ax[0])
sns.barplot(data = data, x = names[2], y = names[1], palette = palette, ax = ax[1])

# For Pt bulk 9.43e6 S/m ==> 9.43e4 S/cm. Probably by increase in grain size
# data
```

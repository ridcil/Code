```python
# Potentiostatic
sample = 'LMO 90 nm N_01'
path = r'C:\Users\lopezb41\OneDrive - imec\Documents\Experiments\Data\Electrochemical\Galvanostatic Deposition\Mimic Round3\Potentiostatic\N_01.txt'
df = pd.read_csv(path, '\t', names = col2, skiprows = 1)
df['Sample'] = sample
df['Current Density (mA/cm$^2$)'] = 1000 * df[col[2]] / area
df['Potential vs Li$^+$/Li (V)'] = df[col[1]] + 3.05 + 0.22

fig, ax = plt.subplots(dpi = 200)
# # ax2 = ax.twinx()
sns.lineplot(data = df, x = col[0], y = 'Current Density (mA/cm$^2$)', ax = ax, palette = pal)
```

```python
#Galvanostatic N_02 0.03 mA/cm2
sample = 'LMO 90 nm N_02'
path = r'C:\Users\lopezb41\OneDrive - imec\Documents\Experiments\Data\Electrochemical\Galvanostatic Deposition\Mimic Round3\Galvanostatic\N_02.txt'

df = pd.read_csv(path, '\t', names = col, skiprows = 1)
df['Sample'] = sample
df['Potential vs Li$^+$/Li (V)'] = df[col[1]] + 3.05 + 0.22

fig, ax = plt.subplots(dpi = 200)
ax2 = ax.twinx()
sns.lineplot(data = df, x = col[0], y = col[1], ax = ax)
sns.lineplot(data = df, x = col[0], y = 'Potential vs Li$^+$/Li (V)', ax = ax2)
```

```python
# CV
path = r'C:\Users\lopezb41\OneDrive - imec\Documents\Experiments\Data\Electrochemical\Galvanostatic Deposition\Mimic Round3\CV'
files = [os.path.join(path, i) for i in os.listdir(path)]
c = [col[0], col[1], col[2], 'Scan']
cv = pd.DataFrame()
for i in files:
    df_0 = pd.read_csv(i, '\t', skiprows=1, names = c)
    cv = pd.concat([cv, df_0], ignore_index=True)
cv['Potential vs Li$^+$/Li (V)'] = cv[col[1]] + 3.05 + 0.22
cv['Current Density ($\mu$A/cm$^2$)'] = 1e6 * cv[col[2]] / area

fig, ax = plt.subplots(dpi = 200)
ax2 = ax.twiny()
sns.scatterplot(data = cv, x = c[1], y = 'Current Density ($\mu$A/cm$^2$)', edgecolor = None, s = 3, ax = ax)
sns.scatterplot(data = cv, x = 'Potential vs Li$^+$/Li (V)', y = 'Current Density ($\mu$A/cm$^2$)', edgecolor = None, s = 2, ax = ax2)
```

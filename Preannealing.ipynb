```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from preannealing_pt import Ec
```

```python
# =============================== LMOref NA Pt ==============================================================================================
path_ref = r'C:\Users\lopezb41\OneDrive - imec\Documents\Experiments\Data\Electrochemical\LSB_09\0I_20'
sample = 'LMOref not annealed Pt 0I_20'
df_cv_lmo_ref_na, capacity_lmo_ref_na, lith_lmo_ref_na, delith_lmo_ref_na, efficiency_lmo_ref_na = Ec.Electrochem(path_ref, 90, 0.63, sample)
data = [df_cv_lmo_ref_na, capacity_lmo_ref_na, lith_lmo_ref_na, delith_lmo_ref_na]
for i in data:
    i['Sample'] = sample


fig, ax = plt.subplots(1, 4, dpi = 100, figsize = (25, 5))
# ax2 = ax[2].twinx()
sns.scatterplot(data = df_cv_lmo_ref_na, x = 'Potential vs Li$^+$/Li (V)', y = 'Current Density ($\mu$A/cm$^2$)', hue = 'CV', ax = ax[0], edgecolor = None, s = 3)
sns.lineplot(data = lith_lmo_ref_na, x = 'Capacity (mAh/cm$^3$)', y = 'Potential vs Li$^+$/Li (V)', ax = ax[1], hue = 'Cycle', palette = 'BuPu')
sns.lineplot(data = delith_lmo_ref_na, x = 'Capacity (mAh/cm$^3$)', y = 'Potential vs Li$^+$/Li (V)', ax = ax[1], hue = 'Cycle', palette = 'BuPu', legend = False)
# ax[1].legend(frameon=False)
sns.scatterplot(data = efficiency_lmo_ref_na[1:], x = 'Cycle' , y = 'Efficiency (%)', ax = ax[3],  c = 'tab:red', s = 100)
sns.scatterplot(data = capacity_lmo_ref_na[1:], x = 'Cycle', y = 'Capacity (mAh/cm$^3$)', hue = 'Charge/Discharge', style = 'Charge/Discharge', ax = ax[2], legend = True, markers = ['s', '^' ], s = 100)
# ax[2].legend(['Efficiency'], loc = 10, frameon = False)
# ax2.legend(frameon = False, loc = 7)
plt.suptitle(sample)
```

```python
sample = 'LMO Annealed Pt 1'

df_charge= pd.concat([capacity_lmo_a1[capacity_lmo_a1['Charge/Discharge'] == 'Charge'], capacity_lmo_a1_2[capacity_lmo_a1['Charge/Discharge'] == 'Charge'], capacity_lmo_a1_3[capacity_lmo_a1['Charge/Discharge'] == 'Charge']])
df_charge['Cycle'] = np.arange(len(df_charge)) + 1
df_discharge= pd.concat([capacity_lmo_a1[capacity_lmo_a1['Charge/Discharge'] == 'Discharge'], capacity_lmo_a1_2[capacity_lmo_a1['Charge/Discharge'] == 'Discharge'], capacity_lmo_a1_3[capacity_lmo_a1['Charge/Discharge'] == 'Discharge']])
df_discharge['Cycle'] = np.arange(len(df_discharge)) + 1
df = pd.concat([df_charge, df_discharge], ignore_index=True)

ef2 = efficiency_lmo_a1_2.copy()
ef3 = efficiency_lmo_a1_3.copy()
ef2['Cycle'] = efficiency_lmo_a1_2['Cycle'] + 5
ef3['Cycle'] = efficiency_lmo_a1_3['Cycle'] + 10
eff = pd.concat([efficiency_lmo_a1, ef2, ef3])
eff['Sample'] = sample

lla2 = lith_lmo_a1_2.copy()
lla3 = lith_lmo_a1_3.copy()
dla2 = delith_lmo_a1_2.copy()
dla3 = delith_lmo_a1_3.copy()
lla2['Cycle'] = lith_lmo_a1_2['Cycle'] + 5
lla3['Cycle'] = lith_lmo_a1_3['Cycle'] + 10
dla2['Cycle'] = delith_lmo_a1_2['Cycle'] + 5
dla3['Cycle'] = delith_lmo_a1_3['Cycle'] + 10

l = pd.concat([lith_lmo_a1, lla2, lla3], ignore_index=True)
d = pd.concat([delith_lmo_a1, dla2, dla3], ignore_index=True)
l['Sample'] = sample
d['Sample'] = sample

# ================== PLOT ========================================

fig, ax = plt.subplots( 1, 3, dpi = 100, figsize = (20,5))
sns.scatterplot(data = df, x = 'Cycle', y = 'Capacity (mAh/cm$^3$)', hue = 'Charge/Discharge', style = 'Charge/Discharge', legend = True, markers = ['s', '^' ], s = 200, ax = ax[0])
ax[0].set_ylim(500, 600)
sns.lineplot(data = l, x = 'Capacity (mAh/cm$^3$)', y = 'Potential vs Li$^+$/Li (V)', hue = 'Cycle', palette = 'BuPu', legend = False, ax = ax[1])
sns.lineplot(data = d, x = 'Capacity (mAh/cm$^3$)', y = 'Potential vs Li$^+$/Li (V)', hue = 'Cycle', palette = 'BuPu', legend = False, ax = ax[1])
sns.scatterplot(data = eff, x = 'Cycle', y = 'Efficiency (%)', s = 150, ax = ax[2])
# ax[2].set_title('Efficiency')
ax[2].set_ylim(90,100)
plt.suptitle(sample)
```

```python
# ======================= COMPARE =======================================

cv = pd.concat([df_cv_lmo_ref_na, df_cv_lmo_a1], ignore_index=True)

lithiation = pd.concat([lith_lmo_ref_na, l], ignore_index=True)

# capacity_all = pd.concat([], ignore_index=True) 

delithiation = pd.concat([delith_lmo_ref_na, d], ignore_index=True) 

efficiency = pd.concat([efficiency_lmo_ref_na, eff], ignore_index= True)

# ========== PLOT CV ================================================================================
fig, ax = plt.subplots(1, 3, figsize = (20,5), dpi = 100)
sns.scatterplot(data = cv, x = 'Potential vs Li$^+$/Li (V)', y = 'Current Density ($\mu$A/cm$^2$)', hue = 'Sample', edgecolor = None, s = 1, ax = ax[0], palette='BuPu')
# ax.legend(bbox_to_anchor=(1,1), frameon = False)
ax[0].set_xlim(3.3, 4.5)

# ========== PLOT lith ================================================================================

sns.scatterplot(data = lithiation, x = 'Capacity (mAh/cm$^3$)', y = 'Potential vs Li$^+$/Li (V)', hue = 'Sample', edgecolor = None, s = 3, ax = ax[1], palette='BuPu')
sns.scatterplot(data = delithiation, x = 'Capacity (mAh/cm$^3$)', y = 'Potential vs Li$^+$/Li (V)', hue = 'Sample', edgecolor = None, s = 3, ax = ax[1], palette='BuPu', legend = False)

sns.scatterplot(data = efficiency[1:], x = 'Cycle',  y = 'Efficiency (%)', ax = ax[2], hue = 'Sample', style = 'Sample', palette='BuPu', s = 100)
ax[2].set_ylim(60,100)
```

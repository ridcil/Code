```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
from lmfit import Model
```

```python
# Model

def line(x, m, b):
    return x * m + b

line_mod = Model(line)
line_params = line_mod.make_params(m = 1, b = 1)

def linear_fit(y, x):
    result = line_mod.fit(y, line_params, x = x)
    # print(result.fit_report())
    print('Slope = ' + str(result.values['m']))
    print('Intercept = ' + str(result.values['b']))
    return result
```

```python
time = np.array([2, 8]) 
thickness = np.array([90, 350])

fit = linear_fit(thickness, time)

fig, ax = plt.subplots(dpi = 200)
plt.scatter(time, thickness)
plt.plot(time, fit.best_fit)
plt.xlabel('Time (h)')
plt.ylabel('Thickness (nm)')
```
